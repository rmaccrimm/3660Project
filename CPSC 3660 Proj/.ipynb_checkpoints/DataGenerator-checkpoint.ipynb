```python
import random
import string

# selects only first word on every line

#read = open('dist-all-last.txt', 'r')

#write = open('last-cleaned.txt','w')

#for line in read:
#    write.write(line.split(None, 1)[0] + '\n')


#returns random line from any passed file
def randomSelect(file):
    f = open(file,'r')
    choice = random.choice(list(f))
    f.close()
    return choice[:len(choice) - 1]

#returns a random zip
def randomZip():
    code = ""
    for i in range(0,3):
        code += random.choice(string.ascii_uppercase) + str(random.randrange(0, 10))
        
    return code

#returns a random street address
def randomAddress(file):
    f = open(file,'r')
    street = random.choice(list(f))
    f.close()
    street = street[:len(street) - 1] + ' '
    return str(random.randrange(1,1000)) + ' ' + street + random.choice(["Road", "Way", "Street", "Boulevard", "Avenue", "Drive"])

# returns a random date
def randomDate(start, end):
    return (str(random.randrange(start, end))
            + '-' + '{:0>2}'.format(random.randrange(1, 13))
            + '-' + '{:0>2}'.format(random.randrange(1, 30)))

#writes passed amount of lines of customer data to a passed filename 
def writeCustomerData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += randomSelect('last-cleaned.txt') + ',' 
        line += randomSelect('first-cleaned.txt') + ',' 
        line += randomAddress('allstreets.txt') + ','
        line += randomSelect('city_names.txt') + ',' 
        line += randomSelect('states.txt') + ',' 
        line += randomZip() + ',' 
        line += str(random.randrange(100000000,1000000000)) + ',' 
        line += randomDate(1930, 2000) + ',' 
        line += str(random.randrange(100000000,1000000000)) + ',' 
        line += str(random.randrange(0, 5)) + ',' 
        line += str(random.randrange(0, 5)) + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing customer data"


#writes passed amount of lines of employee data to a passed filename 
def writeEmployeeData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += randomSelect('last-cleaned.txt') + ',' 
        line += randomSelect('first-cleaned.txt') + ',' 
        line += str(random.randrange(100000000,1000000000)) + ','
        line += str(random.randrange(0,10)) + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing employee data."
    
    
#writes passed amount of lines of employee job data to a passed filename 
def writeEmployeeJobData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += "JobTitle" + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing employee job data."

    
#writes passed amount of lines of employment history data to a passed filename 
def writeEmploymentHistoryData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(random.randrange(0, 100)) + ','
        line += randomSelect('business-names.txt') + ','
        line += random.choice(["Janitor", "Supervisor", "Lackey", "Goon", "CEO", "Employee", "Data Analyst", "Pipe Fitter", "Accountant"]) + ',' 
        line += randomSelect('first-cleaned.txt') + ' ' + randomSelect('last-cleaned.txt') + ','
        line += str(random.randrange(100000000,1000000000)) + ','
        line += randomAddress('allstreets.txt') + ','
        line += randomDate(1930, 2017) + ','
        line += str(i) + '\n'
        
        
        data.write(line)
    
    data.close()
    
    print "Done writing employment history data."

#writes passed amount of lines of payment data to a passed filename 
def writePaymentData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line = str(random.randrange(0, 100)) + ','
        line += randomDate(1999, 2018) + ','
        line += randomDate(1999, 2018) + ','
        line += randomDate(1999, 2018) + ','
        line += str(round(random.uniform(20,2000), 2)) + ','
        line += str(random.randrange(100000000,1000000000)) + '\n' 
    
        data.write(line)
    
    data.close()
    
    print "Done writing payment data."
    
    
#writes passed amount of lines of purchase data to a passed filename 
def writePurchaseData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = randomDate(1999, 2018) + ','
        line += randomSelect('city_names.txt') + ','
        line += random.choice(['0', '1']) + ','
        line += str(random.randrange(0, 100)) + ','
        line += str(random.randrange(0, 100)) + ','
        line += str(i) + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing purchase data."
    
    
#writes passed amount of lines of sales data to a passed filename 
def writeSalesData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += str(random.randrange(0, 100)) + ','
        line += str(random.randrange(0, 100)) + ','
        line += randomDate(1999, 2018) + ','
        line += str(round(random.uniform(0,100), 4)) + ','
        line += str(round(random.uniform(0,100), 4)) + ','
        line += str(round(random.uniform(0,100), 4)) + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing sales data."
    
    
#writes passed amount of lines of sales comission data to a passed filename 
def writeSalesComissionData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(random.randrange(0, 100)) + ','
        line += str(round(random.uniform(100,2000), 2)) + ','
        line += str(i) + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing sales comission data."
    

#writes passed amount of lines of vehicle data to a passed filename 
def writeVehicleData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += str(random.randrange(0, 1000000)) + ','
        line += str(random.randrange(0, 1000)) + ','
        line += str(random.randrange(5000, 1000000)) + ','
        line += str(random.randrange(5000, 1000000)) + ','
        line += random.choice(["SUV", "Truck", "Sedan", "Van", "Coupe", "Wagon", "Convertible", 
                               "Sports Car", "Diesel", "Crossover", "Luxury Car", "Hybrid/Electric", 
                               "Certified Pre-Owned"]) + ','
        line += random.choice(["Red", "Orange", "Yellow", "Green", "Blue", "Indigo", "Purple"]) + ','
        line += random.choice(["Red", "Orange", "Yellow", "Green", "Blue", "Indigo", "Purple"]) + ','
        line += randomSelect('car-makes.txt') + ','
        line += random.choice(["Sonic", "Civic", "Jetta", "Fit", "Rio", "Focus", "Lancer", "Beetle", "Accord"]) + ','
        line += str(random.randrange(1950, 2018)) + '\n'
    
        data.write(line)
    
    data.close()
    
    print "Done writing vehicle data."
    
    
#writes passed amount of lines of vehicle owner data to a passed filename 
def writeVehicleOwnerData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(random.randrange(0, 100)) + ','
        line += str(random.randrange(0, 100)) + '\n'
        
        data.write(line)
    
    data.close()
    
    print "Done writing vehicle owner data."
    
    
#writes passed amount of lines of vehicle problem data to a passed filename 
def writeVehicleProblemData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += "BLAH BLAH BLAH BLAH BLAH"
        line += str(random.randrange(0, 100)) + ','
        line += str(random.randrange(5000, 1000000)) + ','
        line += str(random.randrange(5000, 1000000)) + '\n'
        
        data.write(line)
    
    data.close()
    
    print "Done writing vehicle problem data."

    
#writes passed amount of lines of warranty data to a passed filename 
def writeWarrantyData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(random.randrange(0, 100)) + ','
        line += str(random.randrange(0, 100)) + ','
        line += str(random.randrange(0, 100)) + ','
        line += str(i) + ','
        line += randomDate(1999, 2018) + ','
        line += str(round(random.uniform(0,100), 4)) + ','
        line += str(round(random.uniform(0,100), 4)) + '\n'
        
        data.write(line)
    
    data.close()
    
    print "Done writing warranty data."

    
#writes passed amount of lines of warranty item data to a passed filename 
def writeWarrantyItemData(filename, lines):
    data = open(filename, 'w')

    for i in range(lines):
        line = str(i) + ','
        line += str(random.randrange(0, 100)) + ','
        line += "THING"
        line += randomDate(1999, 2018) + ','
        line += str(random.randrange(1, 30)) + ','
        line += str(round(random.uniform(0,100), 4)) + ','
        line += str(round(random.uniform(0,100), 4)) + '\n'
        
        data.write(line)
    
    data.close()
    
    print "Done writing warranty item data."

writeCustomerData('customerData.txt', 20)
writeEmployeeData('employeeData.txt', 20)
writeEmployeeJobData('employeeJobData.txt', 20)
writeEmploymentHistoryData('employmentHistoryData.txt', 20)
writePaymentData('paymentData.txt', 20)
writePurchaseData('purchaseData.txt', 20)
writeSalesData('salesData.txt', 20)
writeSalesComissionData('salesComissionData.txt', 20)
writeVehicleData('vehicleData.txt', 20)
writeVehicleOwnerData('vehicleOwnerData.txt', 20)
writeVehicleProblemData('vehicleProblemData.txt', 20)
writeWarrantyData('warrantyData.txt', 20)
writeWarrantyItemData('warrantyItemData.txt', 20)
```
